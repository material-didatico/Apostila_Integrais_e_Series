# Método de Newton como uma ilustração das aplicações de Sequências Numéricas


Apresentamos aqui uma implementação do Método de Newton para ilustrar as aplicações das Sequências Numéricas. Ela não é uma implementação eficiente e robusta do método.

O Método de Newton é usado para encontrar zeros de funções deriváveis,
$f(x^\ast)=0$, através de um processo iterativo partindo de um chute inicial $x_0$ e refinando a estimativa pela fórmula
\begin{equation*}
  x_{n+1} = x_n - \frac{\,f(x_n)\,}{\,f'(x_n)\,}
\end{equation*}

## Implementação do Método ne Newton


Definimos uma função que implementa o Método de Newton e exibe graficamente as sequências numéricas geradas.

Parâmetros:
1. func - uma string com a expressão da função em $x$
2. Xo   - o chute inicial $x_0$
3. tol  - a tolerância aceitável para o erro e
4. N    - o número máximo de passos

A função converte a string em uma expressão SymPy e calcula sua derivada. Depois aplica o Método de Newton e exibe as sequências numéricas geradas.

```python
from sympy import *
from sympy.plotting import plot 

import numpy as np
import matplotlib.pyplot as plt

init_printing()

def newton_method(func, Xo, tol, N):

  # Usando computação simbólica para calcular a derivada da função
  s_x  = sympify('x')
  s_f  = sympify(func)
  s_df = diff(s_f, s_x)
  
  # Criando funções lambda para avaliação em ponto flutuante
  f  = lambdify(s_x, s_f,  'numpy')
  df = lambdify(s_x, s_df, 'numpy')

  # Criando arrays para armazenar os passos do método
  n   = np.arange(N+1)
  xn  = np.empty (N+1)
  fn  = np.empty (N+1)
  dfn = np.empty (N+1)

  xn [:] = np.nan
  fn [:] = np.nan
  dfn[:] = np.nan

  # Inicialização do Método de Newton
  xn [0] = Xo
  fn [0] = f(Xo)
  dfn[0] = df(Xo)

  for k in range(N):

    xn [k+1] = xn[k] - fn[k] / dfn[k]
    fn [k+1] =  f(xn[k+1])
    dfn[k+1] = df(xn[k+1])

    if abs(fn[k+1]) < tol or abs(dfn[k+1]) < 1e-9:
      break

  # Exbindo a função e as sequências geradas
  # fig, axs = plt.subplots(2, 3, figsize=(15, 9))

  xx = np.linspace( Xo-5, Xo+5)

  # Exibindo a função e sua derivada
  fig = plt.figure(figsize=(15, 9)) 
  
  a = plt.subplot2grid((2, 3), (0, 0), colspan=3)
  a.plot(xx, f(xx), color='magenta')
  a.plot(xn, fn, 'o', color='blue')
  a.axhline(y=0, color='k')
  a.axvline(x=0, color='k')
  a.set_title('f(x)')

  a = plt.subplot2grid((2, 3), (1, 0))
  a.plot(n, xn, linewidth=0.5, color='blue')
  a.plot(n, xn, 'o', color='blue')
  a.grid()
  a.set_title('Sequência $x_n$')

  a = plt.subplot2grid((2, 3), (1, 1))
  a.plot(n, fn, linewidth=0.5, color='blue')
  a.plot(n, fn, 'o', color='blue')
  a.grid()
  a.set_title('Sequência $f(x)_n$')
    
  a = plt.subplot2grid((2, 3), (1, 2))
  a.plot(n, dfn, linewidth=0.5, color='blue')
  a.plot(n, dfn, 'o', color='blue')
  a.grid()
  a.set_title("Sequência $f'(x)_n$")
  
  plt.show()
```

## Aplicação do Método de Newton para encontrar zeros de algumas funções

```python
newton_method('x**3 - 8', 1, 1e-4, 10 )
```

```python
newton_method('sin(x)', 2, 1e-4, 10 )
```
