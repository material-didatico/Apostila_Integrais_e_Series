# Exemplos de Sequências Numéricas 


Neste exemplo definimos uma função em Python que exige o gráfico dos primeiros termos de uma sequência. Depois a usamos para ilustrar algumas sequências.

## Função para exigir o gráfico dos primeiros termos de uma sequência numérica

Esta célula importa os módulos necessários e cria uma função para exibir os primeiros termos da sequência

```python

import numpy as np
import matplotlib.pyplot as plt

def plot_sequence(n, a):

  fig, ax = plt.subplots()
  ax.plot(n, a, linewidth=0.5, color='blue')
  ax.plot(n, a, 'o', color='blue')
  plt.show()
```

## Exemplos de sequências definidas diretamente por expressões envolvendo o índice

A próxima célula define o array com os índices

```python
n = np.arange(1, 17)
```

As próximas células calculam os primeiros termos de cada sequência e exibem seus gráficos

```python
u = np.ones(n.shape)

plot_sequence(n, u)
```

```python
a = u + (-u)**n

plot_sequence(n, a)
```

```python
b = 1 / n

plot_sequence(n, b)
```

```python
c = n**2

plot_sequence(n, c)
```

```python
d = (-u)**n * n**2

plot_sequence(n, d)
```

```python
e = (-u)**n / n**2

plot_sequence(n, e)
```

## Exemplos de sequências definidas por relações de recorrência

Nesses casos precisamos calcular os termos um a um.


**Atenção**: Observe que o valor do índice do array, que sempre começa em zero,
pode não coincidir com o valor índice da sequência. 
Em nossos exemplos o índice das sequências começam em um, portanto, 
precisamos ajustar seus valores. Se k é o índice da sequência então
k-1 é o índice correspondente para o array.

```python
# Criando um array com zeros para armazenar os termos da sequência
f = np.zeros(n.shape)

# Definindo o valor inicial
k = 1
f[k-1] = 1

# Criando um loop para calcular os demais termos
for k in n[1:]:
  f[k-1] = 0.8 * f[k-2]

plot_sequence(n, f)
```

```python
g = np.zeros(n.shape)

k = 1
g[k-1] = 1

k = 2
g[k-1] = 1

for k in n[2:]:
  g[k-1] = g[k-2] + g[k-3]
 
plot_sequence(n, g)
```

```python
h = np.zeros(n.shape)

k = 1
h[k-1] = 1

for k in n[1:]:
  h[k-1] = h[k-2] / k

plot_sequence(n, h)
```
