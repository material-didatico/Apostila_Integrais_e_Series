# Exemplos de Sequências Numéricas - O Mapa Logístico

O [Mapa Logístico](https://pt.wikipedia.org/wiki/Mapa_log%C3%ADstico) é uma relação de recorrência definida pela regra
\begin{equation*}
  x_{n+1} = r x_n \left( 1 - x_n \right)
\end{equation*}

O comportamento dessa sequência depende do valor da constante $r$.

## Função para calcular e exigir os primeiros termos do mapa logístico


```python

import numpy as np
import matplotlib.pyplot as plt

def plot_mapa_logistico(Xo, r, N):

  x = np.zeros(N)
  x[0] = Xo

  for k in range(1,N):
    x[k] = r * x[k-1] * ( 1 - x[k-1] )

  fig, ax = plt.subplots(figsize=(15,6))
  ax.plot(x, linewidth=0.2, color='blue')
  ax.plot(x, 'o', color='blue')
  plt.show()
```

## Exemplos de mapas logísticos

```python
plot_mapa_logistico( 0.01, 1.5, 31 )
```

```python
plot_mapa_logistico( 0.01, 2.9, 31 )
```

```python
plot_mapa_logistico( 0.01, 3, 31 )
```

```python
plot_mapa_logistico( 0.01, 3.7, 31 )
```
